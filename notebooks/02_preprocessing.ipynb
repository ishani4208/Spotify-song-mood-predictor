```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle

from google.colab import files
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv('spotify_with_moods.csv')
print(f"Shape: {df.shape}")
df.head()
```

```python
cols_to_drop = ['Unnamed: 0', 'track_id', 'artists', 'album_name', 'track_name', 'track_genre']

df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])
print(f"Shape after dropping: {df.shape}")
print(df.columns.tolist())
```

```python
print("Nulls per column:")
print(df.isnull().sum())

df = df.dropna()
print(f"\nShape after dropping nulls: {df.shape}")
```

```python
FEATURES = [
    'danceability', 'energy', 'valence', 'tempo',
    'acousticness', 'speechiness', 'loudness', 'instrumentalness'
]
TARGET = 'mood'

X = df[FEATURES]
y = df[TARGET]

print(f"X shape: {X.shape}")
print(f"Class distribution:\n{y.value_counts()}")
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Label mapping:", dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded      # keeps class balance in both splits
)

print(f"Train size: {X_train.shape[0]}")
print(f"Test size:  {X_test.shape[0]}")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)   # fit only on train
X_test_scaled  = scaler.transform(X_test)         # apply same scale to test

print("Mean of first feature (should be ~0):", X_train_scaled[:, 0].mean().round(4))
print("Std of first feature (should be ~1):",  X_train_scaled[:, 0].std().round(4))
```

```python
# Save scaler
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# Save label encoder (you'll need this in the app too)
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(le, f)

# Save processed splits as CSV (optional but useful for Day 3)
train_df = pd.DataFrame(X_train_scaled, columns=FEATURES)
train_df['mood'] = y_train
train_df.to_csv('train_data.csv', index=False)

test_df = pd.DataFrame(X_test_scaled, columns=FEATURES)
test_df['mood'] = y_test
test_df.to_csv('test_data.csv', index=False)

print("Saved: scaler.pkl, label_encoder.pkl, train_data.csv, test_data.csv")
```
